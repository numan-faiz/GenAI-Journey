```python
df=sns.load_dataset('tips')
df
```

barplot

Barplot (Bar Chart) aik aesa graph hota hai jo mukhtalif categories (groups) ka aapas me muqabla (comparison) karne ke liye istemaal hota hai. Is me har category ko aik alag bar (dabay) se dikhaya jata hai.

```python
sns.barplot(x=df['sex'],y=df['total_bill'])
```

```python
import numpy as np
sns.barplot(x=df['sex'],y=df['total_bill'],estimator=np.sum)
```

boxplot

Boxplot (jise Box-and-Whisker plot bhi kehte hain) aik aesa graph hota hai jo aapke data ki summary (khulasa) dikhata hai aur yeh batata hai ke data kahan zyadah phaila hua hai aur kahan outliers (ajeeb ya bohot alag values) hain.

```python
sns.boxplot(x='tip',y='day',data=df,palette='rainbow')
```

violinplot

Violin plot aik aesa graph hai jo Boxplot aur KDE Plot (Density Curve) ko aapas me mila kar banaya gaya hai.Yeh dekhne me aik "Violin" (baja) ya dholak jaisa lagta hai. Yeh aapko data ki summary bhi deta hai aur yeh bhi dikhata hai ke data kahan zyadah jama hai.

```python
sns.violinplot(x='day',y='total_bill',data=df)
```

striplot

Stripplot aik aesa graph hota hai jo data ke har aik single point ko axis par aik dot (nukte) ki shakal me laga kar dikhata hai [1].Yeh asal me aik Scatter Plot ki tarah hota hai, lekin yeh tab use hota hai jab aapka aik variable category (jaise mard/aurat, ya dukan A/dukan B) ho aur doosra variable continuous (jaise umar ya paise) ho [1]

```python
sns.stripplot(x='day',y='total_bill',data=df)
```

swarmplot

Swarmplot bilkul stripplot ki tarah hota hai, lekin isme data ke dots aik doosre ke upar overlap (chupte) nahi hain. Yeh dots ko automatically daayein-baayein (left-right) phaila kar aik "bee swarm" (shehed ki makhion ke jhund) jaisi shakal bana deta hai.Isay "Beeswarm plot" bhi kehte hain.

```python
sns.swarmplot(x='day',y='total_bill',data=df)
```

combine violin and swarmplot at same time

```python
sns.violinplot(x='tip',y='day',data=df)
sns.swarmplot(x='tip',y='day',data=df)
```

## plotly and cuffling

1. Plotly (The Engine)
Plotly ek data visualization library hai jiska kaam interactive charts banana hai.

Ye kya karta hai? Normal graphs (jaise Matplotlib ke) static hote hain—aap unhe bas dekh sakte ho. Lekin Plotly ke graphs me aap mouse le ja kar values dekh sakte ho (hover), zoom kar sakte ho, aur data points ko hide/show kar sakte ho.

Kyun use karein? Jab aapko dashboards banane hon ya web browser me chalne wale dhasu aur interactive charts chahiye hon.

2. Cufflinks (The Shortcut)
Cufflinks ek helper library hai jo sirf Pandas ke dataframes ko Plotly ke sath jodne ke liye bani hai.

Ye kya karta hai? Pandas me standard graph banane ke liye hum .plot() likhte hain. Cufflinks aapko ek naya command deta hai: .iplot() (Interactive Plot).

Kyun use karein? Agar aap Pandas use kar rahe ho aur aap Plotly ka lamba syntax nahi likhna chahte, toh Cufflinks background me khud hi aapka data utha kar Plotly ka interactive chart bana deta hai—sirf ek line me.
