make new column fature 

col name= family_type

1-> alone

2-> small

1-> lage

```python
def transform_family(num):
    if num==1:
        return "alone"
    elif num>1 and num<5:
        return "small"
    else:
        return "large"
```

```python
df['Family_type']=df['Family_size'].apply(transform_family)
```
