### 1. CNN Kya Hai aur Kyun Use Hota Hai?

* **Definition:** CNN ek khass kism ka deep neural network hai jo **images (tasveeron)** ke liye bana hai.
* **ANN vs CNN:** Purana ANN har pixel ko alag alag dekhta tha aur pixels ka aapsi rishta (**spatial info**) bhool jata تھا. CNN tasveer ko matrix shakal me hi rakh kar uske aas-paas ke pixels ke jor ko samajhta hai.
* **Faida:** Isse computer ko kam weights (**fewer parameters**) seekhne parte hain aur yeh behtar kaam karta hai.

---

### 2. Main Layers Aur Unka Kaam (The Pipeline)

Ek standard CNN model me yeh 4 main steps hote hain:

1. **Convolution Layer (Features Nikalna):**
* Isme ek chota sa matrix jise **Filter ya Kernel** kehte hain, woh poori image par slide (scan) karta hai.
* Dono matrices ke numbers aapas me multiply aur sum ($\sum \sum$) hote hain.
* Jawab me ek **Feature Map** banta hai jo image ke khass patterns (lines, edges) ko bahaar nikal leta hai.


2. **ReLU Activation (Speed aur Teerpan):**
* Convolution ke foran baad lagta hai. Yeh saari negative values ko $0$ kar deta hai.
* Iska kaam data me **non-linearity** lana hai taake model mushkil patterns seekh sake aur training tez ho.


3. **Pooling Layer (Size Chota Karna):**
* Yeh image ka spatial size chota (**downsampling**) karti hai taake computer par load kam ho aur overfitting na ho.
* **Max Pooling:** Apne dabbe me se sabse badi (maximum) value uthata hai.
* **Average Pooling:** Dabbe ke saare numbers ka average nikalta hai.


4. **Fully Connected & Output Layer (Final Faisla):**
* Aakhir me saare bache hue numbers ko ek seedhi line me flat kar ke simple dense layer me daala jata hai jo final class predict karti hai (e.g., Kutta ya Billi).

### 🎯 Shortest Revision Checklist:

* **Convolution:** Scanning kar ke patterns/edges nikalna.
* **ReLU:** Negative ko 0 karna aur non-linearity lana.
* **Pooling:** Data ka size chota karna aur zaroori cheez rakhna.
* **Spatial Hierarchy:** Pehle choti lines $\rightarrow$ phir shapes $\rightarrow$ aakhir me poora object pehchanna.
