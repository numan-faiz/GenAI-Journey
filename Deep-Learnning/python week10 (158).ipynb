ft=sigma(Wff..[ht-1,xt]+bf)

input gate:

it=sigma(Wi.[ht-1,xt]+bi)

ot=sigma(Wo.[ht-1,xt]+bo)


candidate cell state:

propose new candidate values:

Ct=tanh(wc.[ht-1,xt]+bc)

ct=ft*ct-1+it*ct

computing hidden state:

### 🎯 LSTM aur Gating Mechanism ka Poora Nichor (Summary)

---

### 1. LSTM Kya Hai aur Isne Kya Masala Hal Kiya?

* **Problem:** Simple RNN lambe paragraphs ya lambi series me purani baatein bhool jata tha kyunke uska gradient zero (**Vanishing Gradient**) ho jata tha.
* **Solution:** LSTM (**Long Short-Term Memory**) ne is bimari ka ilaj do cheezon se kiya: Ek lambi muddat ka alag dimaag banaya jise **Cell State** kehte hain, aur dimaag par pehre-dar bitha diye jinhein **Gates (Darwaze)** kehte hain.

---

### 2. LSTM Ke Do Main Dimaag (States)

1. **Cell State ($c_t$ - Long-Term Memory):** Yeh network ke upar ek **Conveyor Belt** ki tarah seedha chalta jata hai. Isme information bina kisi rukawat ke aakhir tak jaa sakti hai, isiliye LSTM lambi baatein nahi bhoolta.
2. **Hidden State ($h_t$ - Short-Term Memory):** Yeh har step par badalta rehta hai aur haliya step ka final output hota hai.

---

### 3. Teeno Gates Ka Kaam (The 3 Gatekeepers)

Saare gates ke piche **Sigmoid ($\sigma$)** hota hai. Agar Sigmoid ka jawab **0** aaye toh darwaza band (Block), aur agar **1** aaye toh darwaza khula (Allow).

* **🟢 Forget Gate ($f_t$):** Yeh faisla karta hai ke pichli memory me se kis faltu baat ko **bhoolna/safaya** karna hai.

$$f_t = \sigma(W_f \cdot [h_{t-1}, x_t] + b_f)$$


* **🔵 Input Gate ($i_t$) & Candidate ($\tilde{C}_t$):** Yeh taye karta hai ke aaj ke naye input me se kaun si kaam ki cheez memory ke andar **jama (add)** karni hai.

$$i_t = \sigma(W_i \cdot [h_{t-1}, x_t] + b_i)$$


* **🔴 Output Gate ($o_t$):** Yeh faisla karta hai ke memory cell ke andar maujood baaton me se kitni baat ko **bahaar output** (Hidden State) me dikhana hai.

$$o_t = \sigma(W_o \cdot [h_{t-1}, x_t] + b_o)$$



---

### 4. Memory Update Kaise Hoti Hai?

Asli Cell State ($c_t$) ko update karne ke liye purani memory ko forget gate se guzarayenge aur naye khayal ko input gate se guzar kar dono ko jama kar denge:

$$c_t = f_t \ast c_{t-1} + i_t \ast \tilde{C}_t$$

---

### 🗂️ Quick Revision Checklist for Quiz:

1. **Cell State:** Long-term conveyor belt (Vanishing gradient se bachata hai).
2. **Sigmoid in Gates:** 0 se 1 ke beech filter karna (0 = Forget/Block, 1 = Keep/Allow).
3. **$\tanh$ in Candidate/Output:** Data ko $-1$ aur $1$ ki range me rakhna taake numbers out of control na hon.
4. **Applications:** Text Translation (NLP), Speech Recognition, aur Stock Market Prediction.
