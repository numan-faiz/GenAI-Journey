CNN-- bulding a CNN Model With tensorflow/keras

CNN architecture overview

input layer:image data (e.g. 28*28*1 for gray scale)

convolution layer+relu activation

pooling layer

repeated convoltion +pooling layers (optional)

flatten layer-->fully connected layers

output layer (softmax for multi-class classification)



```python
import tensorflow as tf
from tensorflow.keras import layers,models
```

```python
mnist=tf.keras.datasets.mnist
(x_train,y_train),(x_test,y_test)=mnist.load_data()

# normalize pixel values
x_train,x_test=x_train/255.0,x_test/255.0

# add channel dimenstion
x_train=x_train[...,tf.newaxis]
x_test=x_test[...,tf.newaxis]
```

```python
model=models.Sequential([
    layers.Conv2D(32,(3,3),activation='relu',input_shape=(28,28,1)),
    layers.MaxPool2D((2,2)),
    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(64,activation='relu'),
    layers.Dense(10,activation='softmax')
])
model.summary()
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(x_train,y_train,epochs=10,validation_split=0.1)
```

```python
test_loss,test_acc=model.evaluate(x_test,y_test)
print(f"test accuracy:,{test_acc:.4f}")
```
