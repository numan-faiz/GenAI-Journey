### 🎯 Training, Padding, aur Stride ka Poora Nichor (Summary)

---

### 1. CNN Training Kya Hoti Hai?

Training ka matlab hai network ko tasveeron par practice karwana taake uska andaza sahi ho sake. Yeh do main steps me hoti hai:

* **Forward Pass:** Tasveer model me aati hai aur model andaza (prediction) lagata hai. Phir check kiya jata hai ke andaza kitna ghalat hai (**Loss Calculation**).
* **Backward Pass (Backpropagation):** Us ghalati (error) ko piche bhej kar model ke weights ko theek kiya jata hai.
* **Batch-wise Training:** Model ek ek tasveer nahi dekhta, balke tasveeron ka gucha (**Batch**, jaise 64 images) ek sath dekh kar apne weights update karta hai.

---

### 2. Padding Kya Hai aur Kyun Zaroori Hai?

* **Kya hai?** Tasveer ke charon taraf $0$ (zeros) ki ek extra boundary lagana.
* **Kyun zaroori hai?** 1. Agar padding na ho, toh tasveer ke konon (**edges**) ki details zaya ho jati hain.
2. Yeh output ka size control karti hai.
* **Types:**
* **Valid Padding:** Koi padding nahi lagti (output size chota ho jata hai).
* **Same Padding:** Itni padding lagti hai ke output ka size bilkul asli input tasveer jitna hi rehta hai.



---

### 3. Stride Kya Hai?

* **Kya hai?** Filter jab tasveer par scan karte hue aage barhta hai, toh uske qadam ka size (**step size**) stride kehlata hai.
* **Asar:** * **Stride = 1:** Filter ek-ek pixel aage sarakta hai (details achhi aati hain, size bara rehta hai).
* **Stride = 2:** Filter do-do pixels kood kar aage jata hai (output size chota ho jata hai aur calculation bach jati hai).



---

### 🧮 Quiz Formula Guide (Sabse Important Math)

Jab bhi Output Matrix ka size nikalna ho, yeh ek hi master formula use hota hai:

$$\text{Output Size} = \left( \frac{\text{Input} - \text{Kernel} + 2 \times \text{Padding}}{\text{Stride}} \right) + 1$$

* **Padding barhanay se** output ka size **bara** hota hai.
* **Stride barhanay se** output ka size **chota** (downsample) hota hai.
