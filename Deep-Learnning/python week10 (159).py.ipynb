# LSTM--building an LSTM model with tensorflowss and keras

tips for better performance:

tune number of LSTM units

adjust learning rate and batch size

use dropout layers to avoid overfiiting

conside stacking multiple LSTM layers for complex data



```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense
import pandas as pd
import numpy as np
```

```python
# laod the data
df_loaded=pd.read_csv('module_157_data.csv')
```

```python
# separate feature and target

x=df_loaded.drop('target',axis=1).values
y=df_loaded['target'].values

# Reshape: from (2000,10) to (2000,10,1)
# this adds the "feature" dimension required by LSTM

x=x.reshape((x.shape[0],x.shape[1],1))
```

```python
# build and compile

model=Sequential([
    LSTM(32,input_shape=(10,1)),
    Dense(1,activation='sigmoid')
])
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# Train

model.fit(x,y,epochs=20,batch_size=32,validation_split=0.2)
```

```python
# evaluate and print the accuracy

print(f"final accuracy:{model.evaluate(x,y,verbose=0)[1]:.4f}")
```
