# RNN -Building a Basic Recurrent neural nerowk

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN,Dense
import pandas as pd
import numpy as np
```

```python
# load data
df_loaded=pd.read_csv('module_157_data.csv')
# separate features and target
x=df_loaded.drop('target',axis=1).values
y=df_loaded['target'].values

# Reshape:from (2000,10) to (2000,10,1)
# this adds the "feature" dimension required by RNN

x=x.reshape(x.shape[0],x.shape[1],1)
```

```python
# build and compile
model=Sequential([
    # simpleRNN is the basic recurrent layers
    SimpleRNN(32,input_shape=(10,1)),Dense(1,activation='sigmoid')

])
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])           #  adative monet estimation
```

```python
# train

model.fit(x,y,epochs=20,batch_size=32,validation_split=0.2)
```

```python
# Evaluate and print only the accuracy
print(f"Final Accuracy: {model.evaluate(x,y,verbose=0)[1]:.4f}")
```
