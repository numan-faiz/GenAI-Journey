### 🎯 RNN aur Vanishing Gradient ka Poora Nichor (Summary)

---

### 1. RNN Kya Hai aur Kyun Use Hota Hai?

* **Sequential Data:** Purane ANN aur CNN sirf static data (jaise alag alag tasveerein) par kaam karte hain. **RNN** ko khass tour par **time-series, text, speech, aur videos** ke liye banaya gaya hai jahan data ek makhsoos tartib (sequence) me aata hai.
* **Memory (Hidden State $h_t$):** Simple networks seedha aage chalte hain, lekin RNN ke andar **loops** hote hain. Iska **Hidden State** ek dimaag ya memory ki tarah kaam karta hai jo pichli baaton ko yaad rakh kar agle step par lekar jata hai.
* **Parameter Sharing:** RNN me sabse khass baat yeh hai ke chahe jitne bhi time steps aa jayein, har step par **wahi purane weights** baar-baar use hote hain (Shared parameters).

---

### 2. RNN Kaam Kaise Karta Hai? (Mathematical Flow)

Har naye step par RNN do cheezon ko aapas me milata hai: **Aaj ka naya input ($x_t$)** aur **pichli memory ($h_{t-1}$)**.

* **Memory Equation:**
$$h_t = \tanh(W_{hh} h_{t-1} + W_{xh} x_t + b_h)$$



*(Purani memory aur naye input ko unke weights se multiply kar ke jama kiya jata hai aur $\tanh$ activation lagaya jata hai).*
* **Output Equation:**
$$y_t = W_{hy} h_t + b_y$$



*(Mojooda memory se final prediction ya output nikala jata hai).*

---

### 3. Vanishing Gradient Problem (RNN ki Sabse Badi Bimari)

* **Bimari Kya Hai?** Jab sequence boht lamba ho (jaise koi bara paragraph ya lambi audio), toh training ke dauran ghalati ko piche bhejte waqt (**Backpropagation Through Time - BPTT**) learning ka ishara (gradient) aahista aahista chota hote hote bilkul khatam ho jata hai.
* **Kyun Hoti Hai?** Kyunke lambe sequences me 1 se chote numbers (derivatives) baar-baar aapas me multiply hote hain, jisse jawab **zero ($0$)** ke qareeb poonch jata hai aur weights update hona band ho jaate hain.
* **Natija (Impact):** Model purani baaton ko bilkul **bhool jata hai** aur sirf haliya (short-term) baatein hi yaad rakh pata hai. Is wajah se simple RNN lambe text ya lambi series par buri tarah flop ho jata hai.

---

### 🗂️ Quick Revision Checklist for Quiz:

1. **ANN/CNN:** Static data (No memory).
2. **RNN:** Sequential data (Has memory via Hidden State).
3. **Weights:** Har time step par bilkul se same rehte hain.
4. **Vanishing Gradient:** Derivatives $<1$ multiply hone se gradient zero ho jata hai, jisse model **long-term dependencies** bhool jata hai.
